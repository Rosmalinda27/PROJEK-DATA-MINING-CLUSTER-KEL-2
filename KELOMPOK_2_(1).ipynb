```python
def optimise_k_means(df1, max_k):
    means = []
    inertias = []

    for k in range(1, max_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(df1)

        means.append(k)
        inertias.append(kmeans.inertia_)

    fig =plt.subplots(figsize=(10,5))
    plt.plot(means, inertias, 'o-')
    plt.xlabel('Number of clusters (K)')
    plt.ylabel('Inertia')
    plt.title('Elbow Method for Optimal K')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
from sklearn.metrics import silhouette_score

# Memilih hanya kolom numerik atau mengonversi kolom kategori menjadi numerik
df_numerik = pd.get_dummies(df1.drop(columns=['Cluster']))

# Mendapatkan label klaster
labels = df1['Cluster'].values

# Evaluasi klasterisasi menggunakan Silhouette Score
sil_score = silhouette_score(df_numerik, labels)

# Menampilkan hasil evaluasi sebagai persentase
print(f'Silhouette Score untuk klasterisasi adalah: {sil_score * 100:.2f}%')
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

silhouette_scores = []

for n_clusters in range(2, 6):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_df1)
    silhouette_avg = silhouette_score(scaled_df1, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {n_clusters}, Silhouette Score = {silhouette_avg}")

max_silhouette_score = max(silhouette_scores)
optimal_num_clusters = silhouette_scores.index(max_silhouette_score) + 2
print(f"\nSilhouette Score yang paling tinggi adalah untuk {optimal_num_clusters} Cluster")
print(f"\nJadi, jumlah Cluster terbaik pada metode Silhouette Coefficient adalah {optimal_num_clusters} Cluster")

plt.plot(range(2, 6), silhouette_scores, marker='o')
plt.title('Silhouette Coefficient for Different Numbers of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Coefficient')
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score

db_indices = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_df1)
    db_index = davies_bouldin_score(scaled_df1, labels)
    db_indices.append(db_index)
    print(f"For n_clusters = {k}, Davies-Bouldin Score = {db_index}")

min_db_indices = min(db_indices)
optimal_num_clusters = db_indices.index(min_db_indices) + 2
print(f"\nDavies-Bouldin Score yang paling rendah adalah untuk {optimal_num_clusters} Cluster")
print(f"\nJadi, jumlah Cluster terbaik pada metode Davies-Bouldin Score adalah {optimal_num_clusters} Cluster")

plt.plot(k_values, db_indices, marker='o')
plt.title('Davies-Bouldin Index for Different Cluster Numbers')
plt.xlabel('Number of Clusters')
plt.ylabel('Davies-Bouldin Index')
plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score

calinski_harabasz_scores = []

# Rentang jumlah cluster yang ingin diuji
for n_clusters in range(2, 6):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_df1)
    ch_score = calinski_harabasz_score(scaled_df1, cluster_labels)
    calinski_harabasz_scores.append(ch_score)
    print(f"For n_clusters = {n_clusters}, Calinski-Harabasz Score = {ch_score}")

max_ch_score = max(calinski_harabasz_scores)
optimal_num_clusters = calinski_harabasz_scores.index(max_ch_score) + 2
print(f"\nCalinski-Harabasz Score yang paling tinggi adalah untuk {optimal_num_clusters} Cluster")
print(f"\nJadi, jumlah Cluster terbaik pada metode Calinski-Harabasz adalah {optimal_num_clusters} Cluster")

# Visualisasi Calinski-Harabasz Scores
plt.plot(range(2, 6), calinski_harabasz_scores, marker='o')
plt.title('Calinski-Harabasz Index for Different Numbers of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Calinski-Harabasz Index')
plt.show()
```
